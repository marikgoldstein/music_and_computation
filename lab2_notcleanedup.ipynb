```python
class Note:
    # n = Note(....)
    def __init__(self,pitch,duration,dynamic='pp',articulation='none'):  
        self.possible_dynamics = ['pp','p','mp','mf','f','ff']
        assert pitch >= 0
        self.pitch = pitch
        self.duration = duration
        assert dynamic in self.possible_dynamics,"invalid dynamic"
        self.dynamic = dynamic
        self.articulation = articulation  
    def halfstep_up(self):
        new = Note(self.pitch+1, self.duration, dynamic=self.dynamic, articulation=self.articulation)
        return new
```

```python
n = Note(40,0.5,dynamic='f',articulation='staccato')
n_higher = n.halfstep_up()
print(n.pitch)
print(n_higher.pitch)
print(n)
print(n_higher)
```

```python
import numpy

rests =['r','rr']
attacks = ['a','aa','aaa']
choices = []

for rest_choice in rests:
    for attack_choice in attacks:
        pair = rest_choice + attack_choice
        choices.append(pair)
#print(choices)

sequence = []
for i in range(40):
    x = numpy.random.choice(choices, size=1)
    sequence.append(x.item())
#print(sequence)
sequence = "".join(sequence)
print(sequence)
```

```python
class Beat:
    def __init__(self,is_note,width=None):
        if is_note:
            assert width is not None
        self.is_note = is_note
        self.width = width
        
    def pretty_print(self):
        if self.is_note:
            print(self.width)
        else:
            print("[]")

widths = ['-','--','---']
```

```python
beats = []

choose = numpy.random.choice

for s in sequence:
    if s == 'r':
        s_is_note = False
        s_width=None
    elif s == 'a':
        s_is_note = True
        s_width = choose(widths,1)
        s_width = s_width.item()
        
    else:
        assert False
        
    beat = Beat(s_is_note,s_width)
    beats.append(beat)
    
for b in beats:
    b.pretty_print()
```
