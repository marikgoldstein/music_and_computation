```python
'''
import some libraries
you can mostly ignore this for now
'''
%matplotlib inline
from scipy.io.wavfile import write
from numpy import linspace,sin,pi,int16
import numpy
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
'''
makes a grid of 44100 evenly spaced
numbers between 0 and 1. This is our "time" axis
for 1 second of sound sampled at 44100 samples pers econd.
'''
length=1
SAMPLE_RATE = 44100
t = numpy.linspace(0,length,length*SAMPLE_RATE)
print(t)
print(len(t))
```

```python
def note(freq=440, length=1, amp=1, rate=SAMPLE_RATE):
    '''
    freq is the frequency of this single pure wave
    length is the length in seconds
    amp is the amplitude
    rate is the sampling rate
    '''
    t = numpy.linspace(0,length,length*rate)
    wav = numpy.sin(2*pi*freq*t)* amp
    return wav

print(note(freq=880,length=2,amp=1,rate=44100))
```

```python
'''
when we hear several frequencies at once,
that's the same as hearing their sum.
Here, sum means each corresponding pair
of samples are added together.
'''
tone1 = note(freq=440,length=10)
tone2 = note(freq=1000,length=10)
result = tone1 + tone2

'''
Some plotting code
'''
plt.title("440hz")
plt.xlabel("Samples (44100 per second)")
plt.ylabel("amplitude")
plt.axis([0, 220, -2, 2])
plt.plot(tone1,'blue',label='440hz')
plt.plot(tone2,'purple',label='1000hz')
plt.plot(result,'red',label='sum')
plt.legend()
plt.show()
```
